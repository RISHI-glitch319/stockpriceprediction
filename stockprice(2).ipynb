```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading csv file
data = pd.read_csv("VTI.csv")
```

```python
data.head()
```

```python
#assigning the data to variables
x = data.drop(['Date','Adj Close','Close'],axis = True)
y = data['Close']
```

```python
x
```

```python
#importing train test split method
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Pair plots for all the data
sns.pairplot(data)
```

```python
#Visulizing the Training plots
plt.plot(x_train,y_train,'r.',label='train_data')
plt.xlabel('x_train')
plt.ylabel('y_train')
plt.legend()
plt.show()
```

```python
#Visualizing the Tesing plots
plt.plot(x_test,y_test,'g.',label='test_data')
plt.xlabel('x_test')
plt.ylabel('y_test')
plt.legend()
plt.show()
```

```python
#import linear regression algorithm
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
#predict the values and assigning
y_predict = regressor.predict(x_test).round(1)
```

```python
calculation = pd.DataFrame(np.c_[y_test, y_predict], columns=['Actual', 'Predicted'])
calculation
```
